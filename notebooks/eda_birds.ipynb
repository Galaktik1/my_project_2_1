```python
import os
import random
from pathlib import Path
from PIL import Image
import matplotlib.pyplot as plt

# Путь к данным
DATA_PATH = Path(r"C:\Users\Galin22\my_project_2_1\data\birds\CUB_200_2011")

# 1. Загрузка классов
with open(DATA_PATH / 'classes.txt') as f:
    classes = f.read().splitlines()

print(f"Классы: {classes}")

# 2. Просмотр изображений
fig, axes = plt.subplots(1, len(classes), figsize=(15, 5))

for i, cls in enumerate(classes):
    img_paths = list((DATA_PATH / 'images').glob(f"*{cls}*.jpg"))
    img_path = random.choice(img_paths)  # случайное изображение
    img = Image.open(img_path)
    axes[i].imshow(img)
    axes[i].set_title(cls)
    axes[i].axis('off')

plt.show()

# 3. Проверка баланса классов
image_class_labels = open(DATA_PATH / 'image_class_labels.txt').read().splitlines()
class_counts = {cls: image_class_labels.count(str(i + 1)) for i, cls in enumerate(classes)}
print("\\nБаланс классов (количество изображений для каждого класса):")
print(class_counts)
```
